```python
import os
import sys
import shutil
import dotenv
from   dotenv import find_dotenv, load_dotenv
import matplotlib
load_dotenv(find_dotenv())
PATH_P     = os.environ['PATH_P']
os.chdir(PATH_P)
sys.path.insert(0, PATH_P)
import pandas as pd
from src.dst.outputhandler.pickle import tic,toc,pickle_save_,pickle_load
from src.dst.plots.plots import plotting_tool,retrieve_data,SeabornFig2Grid
from src.visualization.vis_LSTM_p.main import main_visualize

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
from src.visualization.vis_LSTM_p.main import main_visualize
from src.models.ensemble.model_selection import model_selection
```

```python
path_T = './models/ensemble/RvsNR/True/'
path_F = './models/ensemble/RvsNR/False/'
array  = []
for i in range(1,10):
    dict_ = {}
    
    path_T_data = path_T +'clusters_'+str(i)+'/best/data_best.p'
    path_F_data = path_F +'clusters_'+str(i)+'/best/data_best.p'

    data_T      = pickle_load(path_T_data,None)
    data_F      = pickle_load(path_F_data,None)

    dict_['clusters'] = int(i)
    dict_['random_T'] = data_T
    dict_['random_F'] = data_F
    
    array.append(dict_)
```

```python
fig = plt.figure(figsize=(16, 4))

ax1 = plt.subplot(131)
for dict_ in array:
    ax1.plot(dict_['random_T']['FPR_v'], dict_['random_T']['TPR_v'], label=dict_['clusters'])

plt.xlabel('AUC_v')
plt.ylabel('Amount')
plt.legend()
plt.title('random True')

ax2 = plt.subplot(132)
for dict_ in array:
    ax2.plot(dict_['random_F']['FPR_v'], dict_['random_F']['TPR_v'], label=dict_['clusters'])

plt.xlabel('AUC_v')
plt.ylabel('Amount')
plt.legend()
plt.title('random False')


ax3 = plt.subplot(133)
AUC_v_a_F = []
AUC_v_a_T = []

for dict_ in array:
    AUC_v_a_F.append(dict_['random_F']['AUC_v'])
    AUC_v_a_T.append(dict_['random_T']['AUC_v'])
plt.plot(AUC_v_a_F,label = 'False')
plt.plot(AUC_v_a_T,label = 'True')
plt.xlabel('AUC_v')
plt.ylabel('Amount')
plt.legend()
plt.savefig('./plots/pic/ensemble.png')
```
